```
#!/usr/bin/python
# -*- coding: utf-8 -*-
import numpy 
import numpy as np
def f(X):
  f=[0., 0., 0., 0.]
  f[0]=2*X[0]+5*X[1]-2*X[2]+X[3]-17
  f[1]=(-1)*(X[1]**2)+3*X[2]**2-18
  f[2]=X[2]**3+4*X[0]*X[2]-2*X[3]**2-79
  f[3]=5*X[0]-15*X[1]+X[2]+4*X[3]+25
  return f
eps=1e-10
itmax=100000#
x = np.array([[1.],[1.],[1.],[1.]])#x=np.array([[1.],[2.],[3.],[4.]])
print(x)

n=x.shape[0]
print(n)

#Pradines Jakobio matricos reiksmes apskaiciavimas:

dx=sum(abs(x))*1e-5; #(Jakobio matrica) vadinama matrica, sudaryta iš funkcijos komponenčių dalinių išvestinių:

print(dx)
f0=f(x)
print(f0)
s = (n, n)

A=np.zeros([n,n])

for i in range(n):
  #print(i)
  x1 = np.array([[1.],[1.],[1.],[1.]])  #x1 = np.array([[1.],[2.],[3.],[4.]])
  # neveikia bandant prilyginti x, tiesiog corruptinasi pacio x duomenys
  x1[i]=x1[i]+dx  #print("x1[i]= ",x1)
  f1=f(x1) # print("f1\n",f1)  #print("np\n",(np.subtract(f1,f0))/dx)
  for j in range(n):
    A[j,i]=(np.subtract(f1[j],f0[j]))/dx
  
print(A) 
# Broideno metodo iteracijos:
fi=f(x);         # pradine funkcijos reiksme
print("Pradinė funkcijos reikšmė",fi)
for iii in range(itmax):  #deltax=(-1*A[:,:,None])/f(x)
  deltax = np.linalg.lstsq(-A, f(x),rcond=None)[0]  #print(deltax)
  x=x+deltax  #print(x)
  fi1=f(x)  #print(fi1)  #print(fi)
  d1 = deltax.conj().transpose()#+
  d2 = np.dot(d1,deltax)#+  #print(d2)
  d3 =np.subtract(fi1, fi)#+  #print(d3)
  d33=(d3-np.dot(A,deltax))#+
  d4=np.dot(d33,d1)#+
  d5=np.dot(d1,deltax)
  d6=d4/d5
  A=A+d6 # print(A)  #A=A +  np.linalg.lstsq(d4, d5,rcond=None)  #print(A)  #A=A+(fi1-fi-A*deltax)*deltax'/(deltax'*deltax);

  n1=np.linalg.norm(deltax)#+
  n2=np.linalg.norm(x)#+
  tikslumas=n1/(n2+n1)  #print(tikslumas)  #tikslumas=np.linalg.lstsq(n1, (x+n1),rcond=None)[0]
  print("Iteracija ", iii+1," Tikslumas: ",tikslumas)
  if tikslumas < eps:
    print("sprendinys x =",x.conj().transpose())
    print("funkcijos reiksme f =", f(x))
    break
  elif iii == itmax:
    print("****tikslumas nepasiektas. Paskutinis artinys x =",x)
    print("funkcijos reiksme f =",f(x))
    break
  fi=fi1; #A=A+((fi1-fi-A*deltax)*d1)/(d1*deltax) #ratio = [ai / bi for ai, bi in zip(a, b)] # Lygciu sistemos funkcija 
```
